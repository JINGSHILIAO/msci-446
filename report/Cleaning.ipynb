```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import parsedatetime
from pytz import timezone

pd.set_option('display.max_columns', None)

cal = parsedatetime.Calendar()
#datetime_obj, _ = cal.parseDT(datetimeString="tomorrow", tzinfo=timezone("US/Pacific"))
```

```python
df = pd.read_json('rNormalized.json')
df


```

```python
# Cleanup a bit more
df1 = df.drop(labels=["steamId", "steam.name",'steam.drm_consolidated_NA', 'V10'], axis=1)
df1
```

```python
# Normalize requiredage (0-1)
df2 = df1
df2["steam.required_age"] = df1["steam.required_age"] / df["steam.required_age"].max()
df2
```

```python
df2.columns
```

```python
# one hot encode categories, genres
df_long = df2.explode("steam.categories_description").explode('steam.genres_description')
df_temp = pd.concat([df_long, pd.get_dummies(df_long["steam.categories_description"],prefix="steam.categories_description", prefix_sep='_')], axis=1)
df_temp = pd.concat([df_temp, pd.get_dummies(df_long['steam.genres_description'],prefix='steam.genres_description', prefix_sep='_')], axis=1)
df3 = df_temp.drop(labels=["steam.categories_description", 'steam.genres_description'], axis=1)
len(df3.columns)
```
